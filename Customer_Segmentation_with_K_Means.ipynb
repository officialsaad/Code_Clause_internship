## Choosing *'Annual Income'* and *'Spending Score'* Column in dataframe

```python
# so we don't have missing value get column values
x = df.iloc[:,[3,4]].values
print(x)
```

## Choosing the Number of Clusters

```python
from sklearn.cluster import KMeans
```

```python
# finding the optimum number of cluster by parameter
# WCSS -> with in the cluster sum of squares value

list = []
for i in range(1,11):
  kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
  kmeans.fit(x)
  list.append(kmeans.inertia_)
```

```python
# Ploting the graph to know which has minimum value
import seaborn as sns
```

```python
sns.set()
plt.plot(range(1,11),list)
plt.title('CutOFF point Graph')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# so optimun number of cluster 3 interpreted from graph
```

## Training the k-Means Clustering Model

```python
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42)

# return the label
labels = kmeans.fit_predict(x)
print(labels)
```

## *Visuallizing all the Cluster*

```python
# plotting centroids and all cluster
plt.figure(figsize=(10,10))
plt.scatter(x[labels==0,0], x[labels==0,1], s=30, c='grey', label ='cluster 1')
plt.scatter(x[labels==1,0], x[labels==1,1], s=30, c='green', label ='cluster 2')
plt.scatter(x[labels==2,0], x[labels==2,1], s=30, c='blue', label ='cluster 3')

# all in x[labels == #,0] 0 repesent the annual income which is first column in x
# all 1 in x[labels == #,1] 1 represent the sprending score which is second column i x

# Ploting Centroids
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1], s=60, c='cyan', label='Centroids')
plt.title('Customer Groups')
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.show()
```
